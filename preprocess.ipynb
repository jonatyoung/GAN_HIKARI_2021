```python
import pandas as pd
```

```python
data = pd.read_csv("./data/ALLFLOWMETER_HIKARI2021.csv").drop(columns=["Unnamed: 0.1","Unnamed: 0","uid","originh","responh"])

data = pd.get_dummies(data,columns=["traffic_category"])
```

```python
data.info()
```

```python
for col in data.columns:
    if (data[col].dtype != "float64") and (data[col].dtype != "int64"):
        print(data[col].dtype)
        data = data.drop(columns=[col],axis=1)
```

```python
data.to_csv("./data/raw_data_preprocessed.csv",index=False)
```
